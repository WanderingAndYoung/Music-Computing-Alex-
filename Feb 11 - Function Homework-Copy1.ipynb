```python
#Imports
import numpy as np
import matplotlib.pylab as plt
import librosa
import librosa.display
```

```python
#Definitions
file = 'avm.wav'
sampleRate = 44100
sig , sr = librosa.load(file,mono=True,sr=None)

#Functions
def plotAudio(filename,sampleRate):
    sig , sr = librosa.load(filename,mono=True,sr=sampleRate)
    plt.plot(sig)

#Time on X Axis
time=np.arange(0,sig.shape[0])/sr
plt.plot(time,sig)

#Amplitude on Y Axis
excursion=max(abs(sig))
sigNew = sig / excursion
plt.plot(time,sigNew)

#Plotting labels on graph
xlabel = 'Time (seconds)'
ylabel = 'Amplitude'
plt.plot(time,sigNew)
plt.xlabel(xlabel)
plt.ylabel(ylabel)
```

```python
#First Plotting
plotAudio('avm.wav',44100)
```

```python
#Graph 2 with 441 sample rate
plotAudio('avm.wav',441)
```
